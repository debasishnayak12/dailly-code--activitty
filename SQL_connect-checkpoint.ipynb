```python
def sql_df(table_name):
    cur.execute(f"SELECT * FROM {table_name}")
    data = cur.fetchall()

    columns = [desc[0] for desc in cur.description]

    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
vowel_starting_city=df[df["City"].str.match('^[AEIOU]')]["City"].unique()
vowel_starting_city=sorted(vowel_starting_city)
print(vowel_starting_city)                          #City starting with vowel 
```

```python
#City ending with vowel

def ending_with_vowel(s):
    vowels='aeiou'
    return s[-1] in vowels
```

```python
vowel_ending_city=df[df["City"].apply(ending_with_vowel)]["City"].unique()
vowel_ending_city=sorted(vowel_ending_city)
vowel_ending_city
```

```python
# FIND CIty which not starts with vowels
def not_starting_with_vowel(s):
    vowels='AEIOU'
    return s[0] not in vowels
```

```python
City_without_starting_vowel=df[df["City"].apply(not_starting_with_vowel)]["City"].unique()
City_without_starting_vowel=sorted(City_without_starting_vowel)
City_without_starting_vowel
```
